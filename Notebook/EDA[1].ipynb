```python
import pandas as pd

ruta = "/content/drive/MyDrive/WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(ruta)

print(df.shape)
df.head()
```

```python
df.info()
```

```python
print("Valores nulos por columna:\n", df.isnull().sum())

print("\nDuplicados:", df.duplicated().sum())
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())

df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
df["Churn"].value_counts().plot(kind="bar")
plt.title("Distribución de Churn")
plt.xlabel("Churn")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df.groupby("gender")["Churn"].value_counts(normalize=True).unstack().plot(kind="bar", figsize=(7,5))
plt.title("Churn por Género")
plt.xlabel("Género")
plt.ylabel("Proporción")
plt.show()
```

```python
plt.figure(figsize=(7,5))
df[df["Churn"]=="Yes"]["tenure"].hist(alpha=0.5, label="Churn")
df[df["Churn"]=="No"]["tenure"].hist(alpha=0.5, label="No Churn")
plt.legend()
plt.title("Tenure según Churn")
plt.xlabel("Meses con la compañía")
plt.ylabel("Cantidad")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df[["tenure","MonthlyCharges","TotalCharges"]].corr()
plt.imshow(corr, cmap="viridis", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Matriz de Correlación")
plt.show()
```

```python
df["Contract"].value_counts().plot(kind="bar", figsize=(7,5))
plt.title("Distribución de tipos de contrato")
plt.show()

df.groupby("Contract")["Churn"].value_counts(normalize=True).unstack().plot(kind="bar", figsize=(7,5))
plt.title("Churn por tipo de contrato")
plt.show()
```
